```python
cols_to_convert = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').str.strip()
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').str.strip()
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').str.strip()
```

```python
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df[df['discounted_price'].notnull()]
```

```python
df
```

```python
df['price_diff'] = df['actual_price'] - df['discounted_price']
```

```python
top_categories = df['category'].value_counts().head(10)
print(top_categories)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Product Categories by Count")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.show()
```

```python
top_discounts = df.sort_values(by='discount_percentage', ascending=False).head(10)
print(top_discounts[['product_name', 'actual_price', 'discounted_price', 'discount_percentage']])
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='discount_percentage', y='product_name', data=top_discounts)
plt.title("Top 10 Products by Discount %")
plt.xlabel("Discount %")
plt.ylabel("Product Name")
plt.show()
```

```python
avg_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10)
print(avg_rating)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=avg_rating.values, y=avg_rating.index)
plt.title("Average Rating by Category")
plt.xlabel("Avg Rating")
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='price_diff', y='rating')
plt.title('Price Difference vs Product Rating')
plt.xlabel('Price Difference (Actual - Discounted)')
plt.ylabel('Rating')
plt.show()
```

```python
df['is_refund_mentioned'] = df['review_content'].str.contains("return|refund|replace", case=False, na=False)

refund_counts = df['is_refund_mentioned'].value_counts()
print(refund_counts)
```

```python
plt.figure(figsize=(6,4))
refund_counts.plot(kind='bar', color=['green', 'red'])
plt.title("Refund/Return Mentions in Reviews")
plt.xticks(ticks=[0,1], labels=['No', 'Yes'], rotation=0)
plt.ylabel("Count")
plt.show()
```

```python
total_sales = df['discounted_price'].sum()
total_orders = df.shape[0]
average_rating = df['rating'].mean()
most_reviewed_product = df.loc[df['rating_count'].idxmax(), 'product_name']
total_categories = df['category'].nunique()
top_category = df['category'].value_counts().idxmax()
```

```python
print("Total Sales Revenue:", total_sales)
print("Total Orders:", total_orders)
print("Average Rating:", average_rating)
print("Most Reviewed Product:", most_reviewed_product)
print("Number of Categories:", total_categories)
print("Top Category:", top_category)
```

```python
top_reviewed = df.sort_values(by='rating_count', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_reviewed, x='rating_count', y='product_name', palette='viridis')
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Rating Count")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
top_categories = df.groupby('category')['discounted_price'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='coolwarm')
plt.title("Top 10 Categories by Total Sales")
plt.xlabel("Total Sales (₹)")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count', 'price_diff']].corr(), 
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
category_reviews = df.groupby('category')['rating_count'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=category_reviews.values, y=category_reviews.index, palette='magma')
plt.title("Top 10 Categories by Total Reviews")
plt.xlabel("Total Reviews")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
top_expensive = df.sort_values('actual_price', ascending=False)[['product_name', 'actual_price']].dropna().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='actual_price', y='product_name', data=top_expensive, palette='viridis')
plt.title("Top 10 Most Expensive Products")
plt.xlabel("Actual Price (₹)")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```
